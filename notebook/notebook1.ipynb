```python
## now as from the data you can see my all three columns are of diffrent test
## now i will create two new columns both the columns will be categprical
## first column of the dataset will be total score of three tests
## and second will be average score of three tests
```

```python
df["total_score"] = df["math_score"] + df["reading_score"] + df["writing_score"]
df["average_score"] = df["total_score"]/3
```

```python
df.drop(columns=["aveareg_score"],axis=1,inplace=True)
```

```python
df.head()
```

```python
## since i know that maximum score in all the three tests is 100
## so i want to know that how many students have scored full marks in tests
reading_full = df[df["reading_score"] == 100]["average_score"].count()
math_full = df[df["math_score"] == 100]["average_score"].count()
writting_full = df[df["writing_score"] ==100]["average_score"].count()
```

```python
print(f"There are {reading_full} students in reading test who has scored full 100 marks")
print(f"There are {math_full} students in math test who has scored full 100 marks")
print(f"There are {writting_full} students in writting test who has scored full 100 marks")
```

```python
## so from above we can observe that there are lesser number of students whcih actually are
## getting full marks in maths as compared to reading and testing test
## whcih usually happens, most students get less marks in maths test
```

```python
## now i will find those students who have scored less than 20 marks for each test
reading_less = df[df["reading_score"] <20]["average_score"].count()
math_less = df[df["math_score"] < 20]["average_score"].count()
writting_less = df[df["writing_score"] <20]["average_score"].count()
```

```python
print(f"There are {reading_less} students in reading test who has scored less than 20 marks out of 100")
print(f"There are {math_less} students in math test who has scored less than 20 marks marks out of 100")
print(f"There are {writting_less} students in writting test who has scored less than 20 marks out of 100")
```

```python
## insights 
## from above insights we get the observation that students have scored worst in maths test
```
